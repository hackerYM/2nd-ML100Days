````python
import os
import numpy as np
import pandas as pd
````

````python
# 設定 data_path

dir_data = '../data/'
````

````python
# 用 pd.read_csv 來讀取資料

f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))

app_train = pd.read_csv(f_app)
print("Data type: {}".format(type(app_train)))
````

````python
# Dataframe information

app_train.info()
````

````python
# Dataframe Statistic

app_train.describe()
````

````python
# Dataframe rows & columns

df_shape = app_train.shape
df_idx = app_train.index
df_col = app_train.columns

print("Dataframe row and column count: {}".format(df_shape))
print("Dataframe describe index:\n    {}".format(df_idx))
print("Dataframe describe column:\n    {}".format(df_col))
````

````python
# pandas.DataFrame.loc: Purely label-location based indexer for selection by label.
# pandas.DataFrame.iloc: Purely label-location based indexer for selection by label.

top3_left3_by_loc = app_train.loc[1:3, 'SK_ID_CURR':'NAME_CONTRACT_TYPE']
top3_left3_by_iloc = app_train.iloc[1:4, 0:3]

print("top3_left3_by_loc:\n{}".format(top3_left3_by_loc))
print()
print("top3_left3_by_iloc:\n{}".format(top3_left3_by_iloc))
````

````python
# Dataframe simple [filter] -- income over 10000000

income_filter = app_train["AMT_INCOME_TOTAL"] > 10E6
app_train[income_filter].loc[:, 'SK_ID_CURR':'AMT_INCOME_TOTAL']
````

````python
# Dataframe simple [where] -- income over 100000

income_filter = app_train["AMT_INCOME_TOTAL"].where(app_train["AMT_INCOME_TOTAL"] > 10E4, other=0)
income_series = pd.Series(income_filter)

income_series[0:10]
````
